# 3.1 浮點數的世界
在AI的世界所有的資料都會使用編碼轉換成浮點數並進行訓練分析
輸入資料>編碼>神經元(模型參數)>解碼>輸出結果
神經網路會透過逐層學習的方式,將一種形式的資料轉換成另一種形式的資料
- 前段神經層:

    物體邊緣檢測or紋理偵測(皮膚,輪廓)
- 後端神經層:

    捕捉更複雜的結構(器官,眼睛,耳朵)
